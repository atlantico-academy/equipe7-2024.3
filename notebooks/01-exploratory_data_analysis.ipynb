## Tratamento de Dados

```python
# Imputar valores para variáveis numéricas (substituindo com a média)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Height'].fillna(df['Height'].mean(), inplace=True)
df['Weight'].fillna(df['Weight'].mean(), inplace=True)
df['CH2O'].fillna(df['CH2O'].mean(), inplace=True)

# Imputar valores para variáveis qualitativas (nominais ou ordinais)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['family_history_with_overweight'].fillna(df['family_history_with_overweight'].mode()[0], inplace=True)
df['SMOKE'].fillna(df['SMOKE'].mode()[0], inplace=True)
df['FAF'].fillna(df['FAF'].mode()[0], inplace=True)
df['TUE'].fillna(df['TUE'].mode()[0], inplace=True)
df['CAEC'].fillna(df['CAEC'].mode()[0], inplace=True)
df['MTRANS'].fillna(df['MTRANS'].mode()[0], inplace=True)

# Verificar se a imputação foi concluída com sucesso
print("\nValores faltantes por coluna após o tratamento:")
print(df.isnull().sum())
```

```python
# Verificar a distribuição das variáveis numéricas depois da imputação
print(df[['Age', 'Height', 'Weight', 'CH2O']].describe())

# Verificar a distribuição das variáveis qualitativas após imputação
print(df[['Gender', 'family_history_with_overweight', 'SMOKE', 'FAF', 'TUE', 'CAEC', 'MTRANS']].value_counts())
```

```python
# Visualizar distribuição antes e depois da imputação (para variáveis numéricas)
# 'Age'

# Visualizar dados antes da imputação
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, color='blue', label='Antes da Imputação')
plt.title('Distribuição de Idade Antes da Imputação')

# Visualizar dados depois da imputação
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Aplicar imputação para este exemplo
plt.subplot(1, 2, 2)
sns.histplot(df['Age'], kde=True, color='green', label='Após Imputação')
plt.title('Distribuição de Idade Após Imputação')

plt.tight_layout()
plt.show()
```

```python
# Visualizar variáveis categóricas antes e depois da imputação
# 'Gender'

# Antes da imputação
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Distribuição de Gênero Antes da Imputação')

# Aplicar imputação e visualizar depois
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)  # Imputação
plt.subplot(1, 2, 2)
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Distribuição de Gênero Após Imputação')

plt.tight_layout()
plt.show()
```

**2.5 Os níveis de atividade física (FAF) influenciam significativamente os níveis de obesidade (NObeyesdad), e essa relação varia de acordo com o gênero?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir o tema e cores
bg_color = "#f4f4f9"  # Cor de fundo da figura
c_palette = "viridis"  # Paleta de cores

# Definir a figura e o layout
fig, axes = plt.subplots(1, 2, figsize=(24, 14), gridspec_kw={'wspace': 0.2, 'hspace': 0.4})

# Definir cor de fundo da figura
fig.set_facecolor(bg_color)

# Scatter plot de Gênero vs Nível de Atividade Física (FAF)
sns.stripplot(data=df, x='Gender', y='FAF', hue='NObeyesdad', dodge=True,
              ax=axes[0], palette=c_palette, alpha=0.6, size=10, jitter=True)
axes[0].set_title("Gênero vs Nível de Atividade Física (FAF) vs Nível de Obesidade")
axes[0].set_xlabel("Gênero")
axes[0].set_ylabel("Nível de Atividade Física (FAF)")
axes[0].grid(True, axis='y', linestyle='--', alpha=0.9)

# Ajustando os limites do eixo Y com uma margem extra
y_min, y_max = df['FAF'].min(), df['FAF'].max()
axes[0].set_ylim(y_min - 0.05 * (y_max - y_min), y_max + 0.05 * (y_max - y_min))

# Scatter plot de Gênero vs Idade
sns.stripplot(data=df, x='Gender', y='Age', hue='NObeyesdad', dodge=True,
              ax=axes[1], palette=c_palette, alpha=0.6, size=10, jitter=True)
axes[1].set_title("Gênero vs Idade vs Nível de Obesidade")
axes[1].set_xlabel("Gênero")
axes[1].set_ylabel("Idade")
axes[1].grid(True, axis='y', linestyle='--', alpha=0.9)

# Ajustando os limites do eixo Y com uma margem extra
y_min, y_max = df['Age'].min(), df['Age'].max()
axes[1].set_ylim(y_min - 0.05 * (y_max - y_min), y_max + 0.05 * (y_max - y_min))

# Exibe o gráfico
plt.show()
```

*A análise dos gráficos revela que o nível de atividade física (FAF) é geralmente mais baixo entre os indivíduos com obesidade severa, especialmente entre os homens, enquanto os indivíduos com obesidade mais leve ou moderada tendem a apresentar maior variação nos níveis de atividade física. Além disso, observa-se uma tendência de aumento dos níveis de obesidade em faixas etárias mais avançadas, com uma prevalência maior entre as mulheres. Esses padrões indicam que a intervenção em atividade física, principalmente para a população mais velha, pode ser uma estratégia importante no controle da obesidade.*

**Hipótese Nula (H0):** Não há diferença significativa entre os níveis de atividade física (FAF) e os níveis de obesidade (NObeyesdad), independentemente do gênero. Ou seja, os níveis de obesidade não são influenciados pela variação nos níveis de atividade física ou pelo gênero.
**Hipótese Alternativa (H1)** Há uma diferença significativa entre os níveis de atividade física (FAF) e os níveis de obesidade (NObeyesdad), dependendo do gênero. Ou seja, os níveis de obesidade variam significativamente com base nos níveis de atividade física e podem apresentar padrões distintos entre homens e mulheres.

*Portanto, a presença de histórico familiar de obesidade pode ser um indicador relevante na identificação de indivíduos em maior risco de obesidade severa, o que pode orientar estratégias de intervenção mais eficazes e personalizadas.*

**3.2 Relação entre Peso e Altura**

*A análise da correlação entre peso e altura, visualizada no gráfico de dispersão, mostra uma relação moderada e positiva entre essas duas variáveis. À medida que a altura aumenta, o peso tende a ser maior, o que é consistente com a expectativa de que indivíduos mais altos, em geral, pesam mais. No entanto, a dispersão dos dados também revela que a relação não é perfeita, com uma ampla variação no peso para pessoas de mesma altura, sugerindo que outros fatores, como composição corporal e hábitos de saúde, também influenciam o peso.*
*Ao analisar a distribuição de diferentes níveis de obesidade (representados pela variável `NObeyesdad`), é possível observar que as pessoas com obesidade mais grave ("Obesity_Type_III") tendem a concentrar-se no lado superior direito do gráfico, indicando maior peso em relação à altura. Já os indivíduos com "Normal_Weight" e níveis mais baixos de obesidade estão mais dispersos ao longo do gráfico, refletindo uma gama maior de pesos para uma determinada altura.*
*Portanto, embora exista uma correlação entre altura e peso, a variabilidade dentro de cada nível de obesidade sugere que a relação não é totalmente preditiva, e que fatores adicionais, como hábitos alimentares, atividades físicas e genética, também desempenham papéis importantes na determinação do peso.*

**3.3 Distribuição do Peso por Classificação de Obesidade**

*A análise dos grupos de peso revela que à medida que se avança na gravidade da obesidade, a dispersão dos dados tende a diminuir (menos variabilidade), com exceção do grupo de obesidade extrema, que apresenta uma dispersão muito alta. Isso sugere que, para indivíduos com obesidade grave, existem características corporais muito variadas que podem refletir diferentes fatores subjacentes (genéticos, metabólicos ou comportamentais). Em contrapartida, para os grupos de peso insuficiente, normal e obesidade moderada (Tipo I), a variação tende a ser mais controlada, o que pode facilitar a aplicação de estratégias de saúde pública ou intervenções específicas.*
*Esses padrões podem ser usados para desenvolver abordagens de saúde mais direcionadas, como estratégias de prevenção para populações em risco de transição para obesidade ou programas de tratamento focados na individualidade do paciente, especialmente para aqueles com obesidade tipo III, onde a variabilidade sugere a necessidade de um tratamento altamente personalizado.*

**3.4 Gênero, Nível de Atividade Física e Nível de Obesidade**

*A análise revela uma tendência de maior obesidade entre mulheres, especialmente aquelas com menores níveis de atividade física. As mulheres, em geral, têm menos probabilidade de se envolver em atividades físicas regulares, o que pode ser um fator contribuidor para o aumento dos níveis de obesidade.*
*Diversos fatores podem influenciar essa baixa taxa de atividade física entre mulheres, incluindo fatores culturais, sociais e de segurança. Além disso, a falta de acesso a espaços adequados para exercício, a sobrecarga de responsabilidades familiares ou até mesmo barreiras psicológicas podem diminuir a motivação para a prática de atividades físicas.*
*Portanto, há uma necessidade urgente de desenvolver programas de saúde e bem-estar focados nas mulheres, que considerem suas necessidades específicas, desde atividades físicas acessíveis até campanhas de conscientização sobre a importância da saúde física. Incentivar a prática regular de atividades físicas pode ser a chave para reduzir os níveis de obesidade, especialmente entre as mulheres.*

**3.5 Idade e Nível de Obesidade**

*A relação entre idade e obesidade sugere que à medida que as pessoas envelhecem, elas tendem a acumular mais peso e a apresentar níveis mais altos de obesidade. Isso pode ser atribuído a uma combinação de fatores, como a diminuição do metabolismo, mudanças hormonais e a redução da atividade física.*
*Com o passar dos anos, o corpo passa por mudanças metabólicas que podem resultar em um aumento do acúmulo de gordura. Além disso, alterações hormonais, como as que ocorrem durante a menopausa nas mulheres, podem contribuir para o aumento da gordura abdominal, que é um indicador de maior risco de doenças cardiovasculares.*
